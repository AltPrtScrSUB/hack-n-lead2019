## Let's focus on: human trafficing (USA)

```python
# Can do search queries for following fields: headline, slug, body, caption, topic, entity, ... (fulltext, main)
import datetime
import re
import pandas as pd

base = datetime.datetime.today()
date_list = [base - datetime.timedelta(days=x) for x in range(30,4000,30)]
date_list=[str(d.year)+'.'+str(d.month)+'.'+str(d.day) for d in date_list]

search_query = 'fulltext:(trafficking OR slave) AND NOT (climate OR doping)'

#for i,date in enumerate(date_list[1:]):
url = 'http://rmb.reuters.com/rmd/rest/json/search?q=' + search_query + '&token=' + authToken + '&format=json' + '&dateRange='+'2019.10.03-2019.11.02'
response = requests.get(url)
jj=json.loads(response.text)
headlines=[]
texts=[]
for j in range(len(jj['results']['result'])):
    if 'World News Summary' not in jj['results']['result'][j]['headline'] and 'NEWS SCHEDULE' not in jj['results']['result'][j]['headline']:
        ids=jj['results']['result'][j]['id']
        url = 'http://rmb.reuters.com/rmd/rest/json/item?id=' + ids + '&token=' + authToken + '&format=json'
        response = requests.get(url)
        try:
            text=re.sub('<p>|\\n|</p>|              |\'s','',json.loads(response.text)['body_xhtml'].strip())
            headlines.append(jj['results']['result'][j]['headline'])
            texts.append(text)
        except:
            print('%.0f was empty'%j)


df=pd.DataFrame(data={'headline':headlines,'text':texts})
df.to_csv('api_reuters.csv',sep='\t', index=False)
```
